 ## Cleaning the data to fit into a ML Model

```python
matches.dtypes
```

```python
matches["Date"] = pd.to_datetime(matches["Date"])
```

## Creating predictors for the ML Model

```python
#La funcion categoriza Home o Away con un valor numerico (1, 0)
matches["Venue_code"] = matches["Venue"].astype("category").cat.codes
```

```python
matches
```

```python
matches["Opp_code"] = matches["Opponent"].astype("category").cat.codes
```

```python
#Borraremos los mins y segs de la hora con una expresion regular y lo convetiremos a int
matches["Hour"] = matches["Time"].str.replace(":.", "", regex = True).astype("int")
```

```python
matches["Day_code"] = matches["Date"].dt.dayofweek
```

```python
#Con un diccionario damos valor al resultado de los partidos 
win_draw_loss_map = {"W": 2, "D": 1, "L": 0}
matches["Target"] = matches["Result"].map(win_draw_loss_map)
```

## Creating the initial ML Model

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
randofor = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = matches[matches["Date"] < '2022-01-01']
test = matches[matches["Date"] > '2022-01-01']
```

```python
predictors = ["Venue_code", "Opp_code", "Hour", "Day_code"]
```

```python
randofor.fit(train[predictors], train["Target"])
```

```python
preds = randofor.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["Target"], preds)
```

```python
acc
```

```python
combined = pd.DataFrame(dict(actual = test["Target"], prediction=preds))
```

```python
pd.crosstab(index = combined["actual"], columns = combined["prediction"])
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(test["Target"], preds, average='micro')
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("Date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset = new_cols)
    return group
```
